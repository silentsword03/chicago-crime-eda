```python
data = pd.read_csv("Crimes_-_2001_to_2021",low_memory=False)
```

```python
## Data cleaning. Finding no. of Null values and percentage
```

```python
print("Percentage of missing values in the columns:\n",(data.isna().sum()/data.shape[0])*100)
```

```python
## Using fillna method to fill the missing values. Using foward fill approach. ffill
```

```python
data = data.fillna(method='ffill') 
```

```python
print('Percentage of missing values after filling::\n',(data.isna().sum()/data.shape[0])*100)
```

```python
data.head(10)
```

```python
print(data.columns)
```

```python
##Drop Location based missing information as it of very less percentile
```

```python
data = data[~data['Latitude'].isna()] 
```

```python
print("Percentage of missing values in the columns after filling:\n",(data.isna().sum()/data.shape[0])*100)
```

```python
## Drop Unique features and unimportant feature,which are not much usefull for the analysis
```

```python
data.drop(['ID', 'Case Number', 'Block',
       'Domestic', 'Beat',
       'Ward', 'Community Area', 'FBI Code', 'X Coordinate',
       'Y Coordinate', 'Year', 'Updated On', 'Latitude', 'Longitude',
       ], axis=1)
```

```python
# convert dates to pandas datetime format
#data.Date = pd.to_datetime(data.Date, format='%m/%d/%Y %I:%M:%S %p')
# setting the index to be the date will help us a lot later on
data.index = pd.DatetimeIndex(data.Date)
```

```python
data.head(10)
```

```python
data.groupby('Primary Type').size().sort_values(ascending=False)
```

```python
data[['X Coordinate', 'Y Coordinate']] = data[['X Coordinate', 'Y Coordinate']].replace(0, np.nan)
data =data.dropna()
```

```python
data.groupby('Primary Type').size().sort_values(ascending=False)
```

```python
plt.figure(figsize=(12,12))
sns.lmplot(x='X Coordinate', y='Y Coordinate', size=10, hue='Primary Type', data=data, fit_reg=False)
plt.ylabel('Longitude', fontsize=12)
plt.xlabel('Latitude', fontsize=12)
plt.show()
```

```python
## where each crime occurs More often
```

```python
topk_loc_descr = data.groupby(['Primary Type', 'Location Description']).size().reset_index(name='counts').groupby('Primary Type').apply(lambda x: x.sort_values('counts',ascending=False).head(3))
```

```python
topk_loc_descr
```

```python
pivotdf = topk_loc_descr.pivot(index='Primary Type', columns='Location Description', values='counts')
fig, ax = plt.subplots(figsize=(12,15))         # Sample figsize in inches
#sns.heatmap(df1.iloc[:, 1:6:], annot=True, linewidths=.5, ax=ax)

sns.heatmap(pivotdf, cmap=plt.get_cmap('jet'),cbar=True,ax=ax)
```

```python
## Crimes occurring per month over the years
```

```python
plt.figure(figsize=(11,5))
data.resample('M').size().plot(legend=False)
plt.title('Number of crimes per month')
plt.xlabel('Months')
plt.ylabel('Number of crimes')
plt.show()
```

```python
data.groupby([data.index.month]).size().plot(kind='barh')
plt.ylabel('Months of the year')
plt.xlabel('Number of crimes')
plt.title('Number of crimes by month of the year')
plt.show()
```

```python
days = ['Monday','Tuesday','Wednesday',  'Thursday', 'Friday', 'Saturday', 'Sunday']
data.groupby([data.index.dayofweek]).size().plot(kind='barh')
plt.ylabel('Days of the week')
plt.yticks(np.arange(7), days)
plt.xlabel('Number of crimes')
plt.title('Number of crimes by day of the week')
plt.show()
```

```python
##Total crimes over the years
```

```python
plt.figure(figsize=(11,4))
data.resample('D').size().rolling(365).sum().plot()
plt.title('Rolling sum of all crimes from 2001 - 2020')
plt.ylabel('Number of crimes')
plt.xlabel('Days')
plt.show()
```
